```python
import pandas as pd
users_ratings=pd.read_csv("test_set_all.csv")
users_pop=pd.read_csv("test_set_pop_only.csv")
```

```python
users=set(users_ratings['userId'].values)
```

```python
len(users)

```

```python
data =pd.DataFrame(columns=['userId','movieIds'])
for user in users:
    movies = users_ratings[users_ratings.userId == user]
    movies = movies.filter(items=['movieId'])
    movies_set = set(movies['movieId'].values)
    
    moviestr = ','.join(str(s) for s in movies_set)
    moviestr
    row = [str(user) , moviestr]
    data.loc[len(data)] = row
```

```python
data.to_csv('exclude_list.tsv',sep='\t',index=False,header=False)
```

```python
pop_movieIds=set(pop_movie['movieId'])
```

```python
import numpy as np
```

```python
user_recs_all['overlapCount']=0
user_recs_pop['overlapCount']=0
user_recs_pop
```

```python
for index,row in user_recs_pop.iterrows():
    movie_string=row["Movies"].split(",")
    movie_string=map(int,movie_string)
    movie_set=set(movie_string)
    count=0
    for x in movie_set:
        if x in pop_movieIds:
            count += 1
    print count
    user_recs_pop.loc[[index],['overlapCount']]=count
```

```python
for index,row in user_recs_all.iterrows():
    movie_string=row["Movies"].split(",")
    movie_string=map(int,movie_string)
    movie_set=set(movie_string)
    count=0
    for x in movie_set:
        if x in pop_movieIds:
            count += 1
    print count
    user_recs_all.loc[[index],['overlapCount']]=count
```
